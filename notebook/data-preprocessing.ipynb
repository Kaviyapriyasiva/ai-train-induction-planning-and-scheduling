import necessary libraries

```python
import pandas as pd
from pathlib import Path
```

Paths

```python

RAW_DIR = Path("../data/raw")
OUTPUT_FILE = Path("../data/processed/processed-data.csv")
```

Loading required attributes

```python


stops = pd.read_csv(
    RAW_DIR / "stops.csv",
    usecols=["stop_id", "stop_name", "stop_lat", "stop_lon"]
)

trips = pd.read_csv(
    RAW_DIR / "trips.csv",
    usecols=["trip_id", "route_id", "service_id", "direction_id"]
)

stop_times = pd.read_csv(
    RAW_DIR / "stop_times.csv",
    usecols=[
        "trip_id",
        "stop_id",
        "arrival_time",
        "departure_time",
        "stop_sequence"
    ]
)

calendar = pd.read_csv(
    RAW_DIR / "calendar.csv",
    usecols=[
        "service_id",
        "monday",
        "tuesday",
        "wednesday",
        "thursday",
        "friday",
        "saturday",
        "sunday"
    ]
)

```

Merging the columns

```python

df = stop_times.merge(stops, on="stop_id", how="left")

df = df.merge(trips, on="trip_id", how="left")


df = df.merge(calendar, on="service_id", how="left")


```

Save the processed data

```python
df.to_csv(OUTPUT_FILE, index=False)
```
